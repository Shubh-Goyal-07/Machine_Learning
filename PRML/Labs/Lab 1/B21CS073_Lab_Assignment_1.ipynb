```python
# vi) Write a function to create a random split of the data into train, validation and test sets in the
# ratio of [70:20:10]. (2 Marks)

def rand_split(data):
  train, val, test = [], [], []
  used = []
  length = len(data)

  test_count = 0
  while test_count< int(length/10):
    ind = random.randint(0, length-1)
    if ind not in used:
      used.append(ind)
      test_count += 1

      test.append(data[ind])

  val_count = 0
  while val_count< int(length/5):
    ind = random.randint(0, length-1)
    if ind not in used:
      used.append(ind)
      val_count += 1

      val.append(data[ind])

  for i in range (0, length):
    if i not in used:
      train.append(data[i])

  return (train, val, test)

datasets = rand_split(data)
train, val, test = datasets[0], datasets[1], datasets[2]

print(len(train), len(val), len(test))
```

# **PROBLEM 5**

```python
from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LogisticRegression as LR
from sklearn.tree import DecisionTreeClassifier as DTC
from sklearn.metrics import confusion_matrix as cm
from sklearn.metrics import precision_score as ps
from sklearn.metrics import recall_score as rs
from sklearn.metrics import f1_score as f1s
from sklearn.metrics import accuracy_score as acc
```

To Disable Convergence Warnings (For Custom Training)

```python
from warnings import simplefilter
from sklearn.exceptions import ConvergenceWarning
simplefilter("ignore", category=ConvergenceWarning)
```

## 1.) Input the Dataset

```python
# Dataset Reference :- https://www.kaggle.com/uciml/breast-cancer-wisconsin-data

data = pd.read_csv("drive/MyDrive/PRML/data.csv")
data
```

## 2.) Convert the String Labels into easily-interpretable Numerics

```python
# Note :- There are many existing Encoders for converting String to Numeric Labels, but for convenience, we used Pandas.

condition_M = data.diagnosis == "M"
condition_B = data.diagnosis == "B"

data.loc[condition_M,"diagnosis"]=0
data.loc[condition_B,"diagnosis"]=1

data
```

## 3.) Converting Dataframe into Numpy Arrays (Features and Labels)

```python
Y = data.diagnosis.to_numpy().astype('int')                                     # Labels

X_data = data.drop(columns=["id","diagnosis","Unnamed: 32"])
X = X_data.to_numpy()                                                           # Input Features
```

## 4.) Splitting the Dataset into Train and Test Portions

```python
user_prompt = 0.3
user_enable = False

x_train,x_test,y_train,y_test = tts(X,Y,test_size=user_prompt,shuffle=user_enable)
```

## 5.) Model Training and Predicting

```python
# Note :- Don't worry about the code snippet here, it is just to produce the predictions for the test data portion of each classifier

logistic_model = LR()
logistic_model.fit(x_train,y_train)
logistic_pred = logistic_model.predict(x_test)

decision_model = DTC()
decision_model.fit(x_train,y_train)
decision_pred = decision_model.predict(x_test)
```

## 6.) Evaluation Metrics (Inbulit v/s Scaratch)

### Confusion Matrix

```python
# Note :- Don't worry about the code snippet here, it is just to produce the predictions for the test data portion of each classifier

logistic_model = LR()
logistic_model.fit(x_train,y_train)
logistic_pred = logistic_model.predict(x_test)

decision_model = DTC()
decision_model.fit(x_train,y_train)
decision_pred = decision_model.predict(x_test)
```

### Average Accuracy

```python
inbuilt_acc_logistic = acc(y_test,logistic_pred)
inbuilt_acc_decision = acc(y_test,decision_pred)

print("Accuracy for Logistic Regression-based Predictions =>",str(inbuilt_acc_logistic*100)+"%")
print("Accuracy for Decision Tree-based Predictions =>",str(inbuilt_acc_decision*100)+"%")

def avg_accuracy(y_test, y_pred):
  precs = 0
  for i in range (0, len(y_test)):
    if y_test[i] == y_pred[i]:
      precs+=1

  return precs/len(y_test)

print()
print("Accuracy for Logistic Regression-based Predictions (Scratch) =>",str(avg_accuracy(y_test, logistic_pred)*100)+"%")
print("Accuracy for Decision Tree-based Predictions (Scratch) =>",str(avg_accuracy(y_test, decision_pred)*100)+"%")
```

### Precision

```python
inbuilt_ps_logistic = ps(y_test,logistic_pred)
inbuilt_ps_decision = ps(y_test,decision_pred)

print("Precision for Logistic Regression-based Predictions =>",str(inbuilt_ps_logistic*100)+"%")
print("Precision for Decision Tree-based Predictions =>",str(inbuilt_ps_decision*100)+"%")

def precision(y_test, y_pred):
  true_p = 0
  false_p = 0
  for i in range (0, len(y_test)):
    if y_pred[i] == 1:
      if y_test[i] == y_pred[i]:
        true_p += 1

      else:
        false_p +=1

  return true_p/(true_p + false_p)

print()
print("Precision for Logistic Regression-based Predictions (Scratch) =>",str(precision(y_test, logistic_pred)*100)+"%")
print("Precision for Decision Tree-based Predictions (Scratch) =>",str(precision(y_test, decision_pred)*100)+"%")
```

### Recall

```python
inbuilt_rs_logistic = rs(y_test,logistic_pred)
inbuilt_rs_decision = rs(y_test,decision_pred)

print("Recall for Logistic Regression-based Predictions =>",str(inbuilt_rs_logistic*100)+"%")
print("Recall for Decision Tree-based Predictions =>",str(inbuilt_rs_decision*100)+"%")

def recall(y_test, y_pred):
  true_p = 0
  false_n = 0
  for i in range (0, len(y_test)):
    if y_pred[i] == 1:
      if y_test[i] == y_pred[i]:
        true_p += 1

    else:
      if y_pred[i]!=y_test[i]:
        false_n +=1

  return true_p/(true_p + false_n)

print()
print("Recall for Logistic Regression-based Predictions (Scratch) =>",str(recall(y_test, logistic_pred)*100)+"%")
print("Recall for Decision Tree-based Predictions (Scratch) =>",str(recall(y_test, decision_pred)*100)+"%")
```

### F-1 Score

```python
inbuilt_f1s_logistic = f1s(y_test,logistic_pred)
inbuilt_f1s_decision = f1s(y_test,decision_pred)

print("F1-Score for Logistic Regression-based Predictions =>",str(inbuilt_f1s_logistic*100)+"%")
print("F1-Score for Decision Tree-based Predictions =>",str(inbuilt_f1s_decision*100)+"%")

def f1_score(y_test, y_pred):
  prec = precision(y_test, y_pred)
  rec = recall(y_test, y_pred)

  f1_score = 2/(1/prec + 1/rec)

  return f1_score

print()
print("F1-Score for Logistic Regression-based Predictions (Scratch) =>",str(f1_score(y_test, logistic_pred)*100)+"%")
print("F1-Score for Decision Tree-based Predictions (Scratch) =>",str(f1_score(y_test, decision_pred)*100)+"%")
```

### Class-Wise Accuracy

```python
def class_accuracy(y_test, y_pred):
  true_p = 0
  false_n = 0
  for i in range (0, len(y_test)):
    if y_pred[i] == 1:
      if y_test[i] == y_pred[i]:
        true_p += 1

    else:
      if y_pred[i]!=y_test[i]:
        false_n +=1

  sens = true_p/(true_p + false_n)

  true_n = 0
  false_p = 0
  for i in range (0, len(y_test)):
    if y_pred[i] == 0:
      if y_test[i] == y_pred[i]:
        true_n += 1

    else:
      if y_pred[i]!=y_test[i]:
        false_p +=1

  spec = true_n/(true_n + false_p)

  return (sens+spec)/2

print("Class Accuracy for Logistic Regression-based Predictions (Scratch) =>",str(class_accuracy(y_test, logistic_pred)*100)+"%")
print("Class Accuracy for Decision Tree-based Predictions (Scratch) =>",str(class_accuracy(y_test, decision_pred)*100)+"%")
```

### Sensitivity

```python
def sensitivity(y_test, y_pred):
  true_p = 0
  false_n = 0
  for i in range (0, len(y_test)):
    if y_pred[i] == 1:
      if y_test[i] == y_pred[i]:
        true_p += 1

    else:
      if y_pred[i]!=y_test[i]:
        false_n +=1

  return true_p/(true_p + false_n)

print("Sensitivity for Logistic Regression-based Predictions (Scratch) =>",str(sensitivity(y_test, logistic_pred)*100)+"%")
print("Sensitivity for Decision Tree-based Predictions (Scratch) =>",str(sensitivity(y_test, decision_pred)*100)+"%")
```
